# Probability Distribution + generating data from randomness

```python
## Normal Distribution
mu, sigma = 0, 0.1 # mean and standard deviation
normal_data = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(normal_data, 500, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
range_normal = np.linspace(min(bins), max(bins), 100)
plt.plot(range_normal, st.norm.pdf(range_normal, mu, sigma), linewidth=2, color='b')
plt.suptitle('Normal Distribution')
plt.show()
```

```python
ount, bins, ignored =  plt.hist(np.random.lognormal(mu, sigma, 1000), 30, normed=True)
range_lognormal = np.linspace(min(bins), max(bins), 10000)
pdf = (np.exp(-(np.log(range_lognormal) - mu) ** 2 / (2 * sigma ** 2)) / (range_lognormal * sigma * np.sqrt(2 * np.pi)))
plt.plot(range_lognormal, pdf, linewidth=2, color='b')
plt.suptitle('Log-Normal Distribution')
plt.show()
```

```python
# uniform 
lo, hi = 0, 100
uniform_data = np.random.uniform(low=lo, high=hi, size=1000)
count, bins, ignored = plt.hist(uniform_data, 30, normed=True)
plt.plot(bins,  bins/bins / (hi - lo), linewidth=2, color='r')
plt.suptitle('Uniform Distribution')
plt.show()
```

```python
a, b = 0.5, 0.5
plt.hist(np.random.beta(a, b, 1000), 30, normed=True)
range_beta = np.linspace(0, 1, 100)
plt.plot(range_beta, st.beta.pdf(range_beta, a, b), linewidth=2, color='b')
plt.suptitle('Beta Distribution')
plt.show()
```

```python
ld = 0.5
exponential_data = np.random.exponential(ld, 1000)
count, bins, ignored = plt.hist(exponential_data, 100, normed=True)
range_exponential = np.linspace(0, max(bins), 100)
plt.plot(range_exponential, st.expon.pdf(range_exponential, 0, ld), color='b')
plt.suptitle('Exponential Distribution')
plt.show()
```

# Scatter plots

# Outliers

```python
print('Drop non-number columns')
batting_numbers = battings.drop(['playerID', 'teamID'], axis=1)
batting_numbers.head()
```

```python
print('Describe data filtered out outliers by | Z-Score | < 3')
cleaned_battings = batting_numbers[(np.abs(st.zscore(batting_numbers)) < 3).all(axis=1)]
cleaned_battings.describe()
```
